```python
from langchain_community.document_loaders import DirectoryLoader

text_directory_loader = DirectoryLoader("data/txtfile/", glob="**/*.txt", loader_cls=TextLoader, loader_kwargs={"encoding": "utf-8"})
text_directory_loader
```
